# Profiling du dataset *Données carroyées issues du dispositif sur les revenus localisés fiscaux et sociaux*

Dans ce notebook, on évalue l'exploitabilité du jeu de données *Données carroyées issues du dispositif sur les revenus localisés fiscaux et sociaux* (https://www.data.gouv.fr/fr/datasets/donnees-carroyees-issues-du-dispositif-sur-les-revenus-localises-fiscaux-et-sociaux-filosofi/) par des algorithmes de Machine Learning. Ce dataset présente beaucoup d’applications possibles, surtout si couplé avec des données cartographiques.

```python
import pandas as pd
import pandas_profiling
from pandas_profiling import ProfileReport
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Profiling des datasets

### Import des jeux de données

**Dataset données carroyés Métropole**

```python
path_metr = "data/Filosofi2015_carreaux_niveau_naturel_metropole.csv" #path csv France metropolitaine
carreaux_metr = pd.read_csv(path_metr)
```

**Dataset données carroyés Martinique**

```python
path_mart = "data/Filosofi2015_carreaux_niveau_naturel_reg02.csv" #path csv Martinique
carreaux_mart = pd.read_csv(path_mart)
```

**Dataset données carroyés La Réunion**

```python
path_reu = "data/Filosofi2015_carreaux_niveau_naturel_reg04.csv" #path csv La Réunion
carreaux_reu = pd.read_csv(path_reu)
```

## Profiling

**Profiling dataset Métropole**

```python
profile_metr = carreaux_metr.profile_report()
#profile_metr
```

**Histogrammes et densités de chaque variable**


```python
fig, axs = plt.subplots(4,10,figsize=(20,10))
for i,col in enumerate(carreaux_metr.columns):
  sns.histplot(carreaux_metr.sample(100), x=col, kde=True, ax=axs[i%4,i//4])
```

**Profiling dataset Martinique**

```python
profile_mart = carreaux_mart.profile_report()
#profile_mart
```

**Histogrammes et densités de chaque colonne**

```python
fig, axs = plt.subplots(4,10,figsize=(20,10))
for i,col in enumerate(carreaux_mart.columns):
  sns.histplot(carreaux_mart.sample(100), x=col, kde=True, ax=axs[i%4,i//4])
```

# Analyse de l'exploitabilité des datasets

- la taille du dataset *Métropole* est adaptée pour faire du Machine Learning(nombre de lignes de l'ordre de $10⁵$ et 31 colonnes); la taille des deux autres datasets est inférieure (nombre de lignes de l'ordre de $10²$), cela pourrait influencer la performance des algorithmes 
- une seule variable catégorielle
- Valeurs manquantes déjà groupées dans des variables *inc*
- Plusieurs variables sont corrélées

Ce dataset est exploitable pour faire du ML et présente beaucoup d’applications possibles, surtout si couplé avec des données cartographiques (disponibles au lien donné plus haut).

Ce dataset pourrait par exemple permettre de répondre à des questions telles que:
peut-on prévoir le revenu moyen d’un ménage français en fonction de sa zone géographique? Quel lien entre la taille du ménage et son indice de pauvreté?
