# Profiling du dataset *Inventaire des gaz à effet de serre*

Dans ce notebook, on évalue l'exploitabilité du jeu de données *Inventaire de gaz à effet de serre territorialisé*(https://www.data.gouv.fr/fr/datasets/inventaire-de-gaz-a-effet-de-serre-territorialise/#_) par des algorithmes de Machine Learning. 

```python
import pandas as pd
import pandas_profiling
from pandas_profiling import ProfileReport
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Import du dataset

```python
url = 'https://www.data.gouv.fr/fr/datasets/r/4072eb84-5093-4490-96e6-bcec87f51ea8'
data_gaz_serre = pd.read_csv(url)
```

## Profiling du dataset

```python
profile_gaz_serre = data_gaz_serre.profile_report()
profile_gaz_serre.to_file("4072eb84-5093-4490-96e6-bcec87f51ea8.html")
```

#### Histogrammes et densités de chaque variable

```python
n_rows = 2
n_cols = 6
fig, axs = plt.subplots(n_rows,n_cols,figsize=(20,10))
for i,col in enumerate(data_gaz_serre.columns):
  sns.histplot(data_gaz_serre.sample(100), x=col, kde=True, ax=axs[i%n_rows,i//n_rows])
```

# Analyse de l'exploitabilité du dataset

- la taille du dataset le rend exploitable pour faire du ML (nombre de lignes de l'ordre de $10^{4}$ et 12 colonnes)
- 10 variables numériques et 2 variables catégorielles
- plusieurs corrélations intéressantes entre variables
- colonne *Autres_transport_international* a plus de 90% des valeurs manquantes

Ce dataset est exploitable pour entraîner des algorithmes de ML, et pourrait permettre de répondre à des questions comme:  quel est le secteur qui émet le plus de gaz à effet de serre en Ile de France? Quelles sont les émissions moyennes liées au secteur des déchets en Auvergne?
